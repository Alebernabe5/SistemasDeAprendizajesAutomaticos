## üü¢ Ejercicio 1: Tu Primera Serie

Crea una Serie de Pandas con los siguientes datos de ventas mensuales (en miles de euros):

## üü° Ejercicio 4: Estad√≠sticas Descriptivas

Crea un DataFrame con datos de ventas diarias de una tienda durante 2 semanas:

```python
np.random.seed(42)
ventas_df = pd.DataFrame({
    'Dia': pd.date_range('2025-01-01', periods=14, freq='D'),
    'Ventas': np.random.randint(100, 300, 14),
    'Clientes': np.random.randint(20, 60, 14),
    'Devoluciones': np.random.randint(0, 15, 14)
})
```

Calcula:
1. Estad√≠sticas descriptivas de todas las columnas num√©ricas (`.describe()`)
2. La venta promedio por cliente (nueva columna 'Ticket_Medio')
3. El d√≠a con mayores ventas
4. El total de ventas en la primera semana vs. la segunda semana

````python
import pandas as pd
import numpy as np
# Crear el DataFrame
np.random.seed(42)
ventas_df = pd.DataFrame({
    'Dia': pd.date_range('2025-01-01', periods=14, freq='D'),
    'Ventas': np.random.randint(100, 300, 14),
    'Clientes': np.random.randint(20, 60, 14),
    'Devoluciones': np.random.randint(0, 15, 14)
})

# Tu c√≥digo aqu√≠
# 1. Estad√≠sticas descriptivas de todas las columnas num√©ricas (`.describe()`)
print(ventas_df.describe())
print("\n")
print("------------------------------------------------")
# 2. La venta promedio por cliente (nueva columna 'Ticket_Medio')
ventas_df['Ticket_Medio'] = ventas_df['Ventas'] / ventas_df['Clientes']
print(ventas_df[['Dia', 'Ticket_Medio']])
print("\n")
print("------------------------------------------------")
# 3. El d√≠a con mayores ventas
dia_max_ventas = ventas_df.loc[ventas_df['Ventas'].idxmax()]
print(dia_max_ventas)
print("\n")
print("------------------------------------------------")
# 4. El total de ventas en la primera semana vs. la segunda semana
primera_semana = ventas_df[0:7]
segunda_semana = ventas_df[7:14]
total_primera_semana = primera_semana['Ventas'].sum()
total_segunda_semana = segunda_semana['Ventas'].sum()
print(f"Total Ventas Primera Semana: {total_primera_semana}")
print(f"Total Ventas Segunda Semana: {total_segunda_semana}")
````

## üü° Ejercicio 5: Manejo de Datos Faltantes

Crea un DataFrame con algunos valores faltantes:

```python
datos_incompletos = pd.DataFrame({
    'Producto': ['A', 'B', 'C', 'D', 'E'],
    'Precio': [10.5, np.nan, 15.0, 12.5, np.nan],
    'Stock': [50, 30, np.nan, 45, 60],
    'Categoria': ['Electr√≥nica', 'Ropa', np.nan, 'Electr√≥nica', 'Ropa']
})
```

Realiza las siguientes tareas:
1. Identifica cu√°ntos valores faltantes hay en cada columna
2. Rellena los precios faltantes con la media de los precios existentes
3. Rellena el stock faltante con 0
4. Rellena la categor√≠a faltante con 'Sin categor√≠a'
5. Verifica que no queden valores faltantes

````python
import pandas as pd
import numpy as np

# Crear el DataFrame
datos_incompletos = pd.DataFrame({
    'Producto': ['A', 'B', 'C', 'D', 'E'],
    'Precio': [10.5, np.nan, 15.0, 12.5, np.nan],
    'Stock': [50, 30, np.nan, 45, 60],
    'Categoria': ['Electr√≥nica', 'Ropa', np.nan, 'Electr√≥nica', 'Ropa']
})

# Tu c√≥digo aqu√≠
# 1. Identifica cu√°ntos valores faltantes hay en cada columna
print(datos_incompletos.isnull().sum())
print("\n")
print("------------------------------------------------")
# 2. Rellena los precios faltantes con la media de los precios existentes
datos_incompletos['Precio'].fillna(datos_incompletos['Precio'].mean(), inplace=True)
# 3. Rellena el stock faltante con 0
datos_incompletos['Stock'].fillna(0, inplace=True)
# 4. Rellena la categor√≠a faltante con 'Sin categor√≠a'
datos_incompletos['Categoria'].fillna('Sin categor√≠a', inplace=True)
# 5. Verifica que no queden valores faltantes
print(datos_incompletos.isnull().sum())
````

## üü° Ejercicio 6: Agrupaci√≥n y Agregaci√≥n (GroupBy)

Crea un DataFrame con datos de ventas por producto y regi√≥n:

```python
ventas_region = pd.DataFrame({
    'Region': ['Norte', 'Sur', 'Este', 'Oeste', 'Norte', 'Sur', 'Este', 'Oeste'],
    'Producto': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B'],
    'Ventas': [120, 95, 135, 110, 85, 140, 100, 125],
    'Unidades': [24, 19, 27, 22, 17, 28, 20, 25]
})
```

Calcula:
1. Las ventas totales por regi√≥n
2. Las ventas promedio por producto
3. El n√∫mero de unidades vendidas por regi√≥n
4. Crea una tabla pivote que muestre ventas por regi√≥n (filas) y producto (columnas)

````python
import pandas as pd
# Crear el DataFrame
ventas_region = pd.DataFrame({
    'Region': ['Norte', 'Sur', 'Este', 'Oeste', 'Norte', 'Sur', 'Este', 'Oeste'],
    'Producto': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B'],
    'Ventas': [120, 95, 135, 110, 85, 140, 100, 125],
    'Unidades': [24, 19, 27, 22, 17, 28, 20, 25]
})

# Tu c√≥digo aqu√≠
# 1. Las ventas totales por regi√≥n
ventas_totales_region = ventas_region.groupby('Region')['Ventas'].sum()
print(ventas_totales_region)
print("\n")
print("------------------------------------------------")
# 2. Las ventas promedio por producto
ventas_promedio_producto = ventas_region.groupby('Producto')['Ventas'].mean()
print(ventas_promedio_producto)
print("\n")
print("------------------------------------------------")
# 3. El n√∫mero de unidades vendidas por regi√≥n
unidades_por_region = ventas_region.groupby('Region')['Unidades'].sum()
print(unidades_por_region)
print("\n")
print("------------------------------------------------")
# 4. Crea una tabla pivote que muestre ventas por regi√≥n (filas) y producto (columnas)
tabla_pivote = ventas_region.pivot_table(values='Ventas', index='Region', columns='Producto', aggfunc='sum')
print(tabla_pivote)
````

## üî¥ Ejercicio 7: Combinaci√≥n de DataFrames (Merge)

Tienes dos DataFrames: uno con informaci√≥n de pedidos y otro con informaci√≥n de clientes.

```python
pedidos = pd.DataFrame({
    'PedidoID': [101, 102, 103, 104, 105],
    'ClienteID': [1, 2, 1, 3, 2],
    'Monto': [250.0, 180.5, 320.0, 95.0, 410.0],
    'Fecha': pd.to_datetime(['2025-01-15', '2025-01-16', '2025-01-16', '2025-01-17', '2025-01-18'])
})

clientes = pd.DataFrame({
    'ClienteID': [1, 2, 3, 4],
    'Nombre': ['Ana Garc√≠a', 'Carlos Ruiz', 'Mar√≠a L√≥pez', 'Juan P√©rez'],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla']
})
```

Realiza:
1. Un merge (inner join) entre pedidos y clientes usando 'ClienteID'
2. Muestra el resultado ordenado por 'Monto' descendente
3. Calcula el monto total de pedidos por ciudad
4. Identifica el cliente que m√°s ha gastado (total de sus pedidos)

````python
import pandas as pd
# Crear los DataFrames
pedidos = pd.DataFrame({
    'PedidoID': [101, 102, 103, 104, 105],
    'ClienteID': [1, 2, 1, 3, 2],
    'Monto': [250.0, 180.5, 320.0, 95.0, 410.0],
    'Fecha': pd.to_datetime(['2025-01-15', '2025-01-16', '2025-01-16', '2025-01-17', '2025-01-18'])
})

clientes = pd.DataFrame({
    'ClienteID': [1, 2, 3, 4],
    'Nombre': ['Ana Garc√≠a', 'Carlos Ruiz', 'Mar√≠a L√≥pez', 'Juan P√©rez'],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla']
})

# Tu c√≥digo aqu√≠

# 1. Un merge (inner join) entre pedidos y clientes usando 'ClienteID'
merged_df = pd.merge(pedidos, clientes, on='ClienteID', how='inner')
print(merged_df)
print("\n")
print("------------------------------------------------")
# 2. Muestra el resultado ordenado por 'Monto' descendente
sorted_df = merged_df.sort_values(by='Monto', ascending=False)
print(sorted_df)
print("\n")
print("------------------------------------------------")
# 3. Calcula el monto total de pedidos por ciudad
monto_total_ciudad = merged_df.groupby('Ciudad')['Monto'].sum()
print(monto_total_ciudad)
print("\n")
print("------------------------------------------------")
# 4. Identifica el cliente que m√°s ha gastado (total de sus pedidos)
cliente_mas_gastador = merged_df.groupby('ClienteID')['Monto'].sum().idxmax()
print(cliente_mas_gastador)
````

## üî¥ Ejercicio 8: An√°lisis de Series Temporales

Crea un DataFrame con datos de temperatura diaria durante enero 2025:

```python
np.random.seed(42)
fechas = pd.date_range('2025-01-01', '2025-01-31', freq='D')
temperaturas = 10 + 5 * np.sin(np.arange(31) * 2 * np.pi / 31) + np.random.normal(0, 2, 31)

df_temp = pd.DataFrame({
    'Fecha': fechas,
    'Temperatura': temperaturas
})
df_temp.set_index('Fecha', inplace=True)
```

Realiza:
1. Calcula la media de temperatura del mes
2. Encuentra los 5 d√≠as m√°s c√°lidos
3. Calcula la temperatura promedio por semana (`.resample('W').mean()`)
4. Cuenta cu√°ntos d√≠as la temperatura estuvo por encima de 12¬∞C

````python
import pandas as pd
import numpy as np
# Crear el DataFrame
np.random.seed(42)
fechas = pd.date_range('2025-01-01', '2025-01-31', freq='D')
temperaturas = 10 + 5 * np.sin(np.arange(31) * 2 * np.pi / 31) + np.random.normal(0, 2, 31)

df_temp = pd.DataFrame({
    'Fecha': fechas,
    'Temperatura': temperaturas
})
df_temp.set_index('Fecha', inplace=True)

# Tu c√≥digo aqu√≠

# 1. Calcula la media de temperatura del mes
media_temperatura = df_temp['Temperatura'].mean()
print(f"Media de temperatura del mes: {media_temperatura}")
print("\n")
print("------------------------------------------------")
# 2. Encuentra los 5 d√≠as m√°s c√°lidos
dias_mas_calidos = df_temp.nlargest(5, 'Temperatura')
print(dias_mas_calidos)
print("\n")
print("------------------------------------------------")
# 3. Calcula la temperatura promedio por semana (`.resample('W').mean()`)
temp_promedio_semana = df_temp.resample('W').mean()
print(temp_promedio_semana)
print("\n")
print("------------------------------------------------")
# 4. Cuenta cu√°ntos d√≠as la temperatura estuvo por encima de 12¬∞C
dias_sobre_12 = df_temp[df_temp['Temperatura'] > 12].shape[0]
print(f"D√≠as con temperatura por encima de 12¬∞C: {dias_sobre_12}")
````

## üî¥ Ejercicio 9: Limpieza y Transformaci√≥n de Datos

Tienes un DataFrame "sucio" con datos de productos:

```python
datos_sucios = pd.DataFrame({
    'producto': ['  laptop  ', 'MOUSE', 'Teclado', 'laptop', 'mouse', '  Monitor'],
    'precio': ['1200‚Ç¨', '25', '45.50‚Ç¨', '1150', '30‚Ç¨', '350'],
    'stock': ['10', '  50  ', '30', '5', 'agotado', '15'],
    'categoria': ['Inform√°tica', 'inform√°tica', 'Inform√°tica', 'Inform√°tica', 'Inform√°tica', 'pantallas']
})
```

Limpia los datos:
1. Elimina espacios en blanco de 'producto' y convi√©rtelo a min√∫sculas
2. Limpia 'precio': elimina '‚Ç¨' y convierte a float
3. En 'stock', reemplaza 'agotado' con 0 y convierte a int
4. Estandariza 'categoria' a "Inform√°tica" o "Pantallas" (capitalizado)
5. Elimina productos duplicados (mismo nombre), manteniendo el de menor precio

````python
import pandas as pd
# Crear el DataFrame
datos_sucios = pd.DataFrame({
    'producto': ['  laptop  ', 'MOUSE', 'Teclado', 'laptop', 'mouse', '  Monitor'],
    'precio': ['1200‚Ç¨', '25', '45.50‚Ç¨', '1150', '30‚Ç¨', '350'],
    'stock': ['10', '  50  ', '30', '5', 'agotado', '15'],
    'categoria': ['Inform√°tica', 'inform√°tica', 'Inform√°tica', 'Inform√°tica', 'Inform√°tica', 'pantallas']
})

# Tu c√≥digo aqu√≠
# 1. Elimina espacios en blanco de 'producto' y convi√©rtelo a min√∫sculas
datos_sucios['producto'] = datos_sucios['producto'].str.strip().str.lower()
print(datos_sucios['producto'])
print("\n")
print("------------------------------------------------")
# 2. Limpia 'precio': elimina '‚Ç¨' y convierte a float
datos_sucios['precio'] = datos_sucios['precio'].str.replace('‚Ç¨', '').astype(float)
print(datos_sucios['precio'])
print("\n")
print("------------------------------------------------")
# 3. En 'stock', reemplaza 'agotado' con 0 y convierte a int
datos_sucios['stock'] = datos_sucios['stock'].str.replace('agotado', '0').astype(int)
print(datos_sucios['stock'])
print("\n")
print("------------------------------------------------")
# 4. Estandariza 'categoria' a "Inform√°tica" o "Pantallas" (capitalizado)
def estandarizar_categoria(categoria):
    if categoria.lower() == 'inform√°tica':
        return 'Inform√°tica'
    elif categoria.lower() == 'pantallas':
        return 'Pantallas'
    else:
        return categoria

datos_sucios['categoria'] = datos_sucios['categoria'].apply(estandarizar_categoria)
print(datos_sucios['categoria'])
print("\n")
print("------------------------------------------------")
# 5. Elimina productos duplicados (mismo nombre), manteniendo el de menor precio
datos_limpio = datos_sucios.drop_duplicates(subset='producto', keep='first')
print(datos_limpio)
````

## üî¥ Ejercicio 10: An√°lisis Completo de Ventas (Caso Real)

Analiza un dataset de ventas de una tienda online:

```python
np.random.seed(42)
n = 100

ventas_completo = pd.DataFrame({
    'Fecha': pd.date_range('2024-01-01', periods=n, freq='D'),
    'Producto': np.random.choice(['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Auriculares'], n),
    'Cantidad': np.random.randint(1, 10, n),
    'Precio_Unitario': np.random.choice([599.99, 19.99, 45.50, 249.99, 79.99], n),
    'Descuento': np.random.choice([0, 0.05, 0.10, 0.15], n),
    'Cliente_Tipo': np.random.choice(['Nuevo', 'Recurrente'], n, p=[0.3, 0.7])
})
```

Realiza un an√°lisis completo:
1. Crea una columna 'Ingreso_Total' = Cantidad √ó Precio_Unitario √ó (1 - Descuento)
2. Calcula el ingreso total por tipo de producto
3. Encuentra el producto m√°s vendido (por cantidad de unidades)
4. Calcula el ticket promedio por tipo de cliente
5. Identifica el mes con mayores ingresos
6. Calcula qu√© porcentaje de las ventas tuvo descuento
7. Crea un resumen ejecutivo con las m√©tricas m√°s importantes

````python
import pandas as pd
import numpy as np
# Crear el DataFrame
np.random.seed(42)
n = 100

ventas_completo = pd.DataFrame({
    'Fecha': pd.date_range('2024-01-01', periods=n, freq='D'),
    'Producto': np.random.choice(['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Auriculares'], n),
    'Cantidad': np.random.randint(1, 10, n),
    'Precio_Unitario': np.random.choice([599.99, 19.99, 45.50, 249.99, 79.99], n),
    'Descuento': np.random.choice([0, 0.05, 0.10, 0.15], n),
    'Cliente_Tipo': np.random.choice(['Nuevo', 'Recurrente'], n, p=[0.3, 0.7])
})

# Tu c√≥digo aqu√≠
# 1. Crea una columna 'Ingreso_Total' = Cantidad √ó Precio_Unitario √ó (1 - Descuento)
ventas_completo['Ingreso_Total'] = ventas_completo['Cantidad'] * ventas_completo['Precio_Unitario'] * (1 - ventas_completo['Descuento'])
print(ventas_completo.head())
print("\n")
print("------------------------------------------------")
# 2. Calcula el ingreso total por tipo de producto
ingreso_total_producto = ventas_completo.groupby('Producto')['Ingreso_Total'].sum()
print(ingreso_total_producto)
print("\n")
print("------------------------------------------------")
# 3. Encuentra el producto m√°s vendido (por cantidad de unidades)
producto_mas_vendido = ventas_completo.groupby('Producto')['Cantidad'].sum().idxmax()
print(f"Producto m√°s vendido: {producto_mas_vendido}")
print("\n")
print("------------------------------------------------")
# 4. Calcula el ticket promedio por tipo de cliente
ticket_promedio_cliente = ventas_completo.groupby('Cliente_Tipo')['Ingreso_Total'].mean()
# 5. Identifica el mes con mayores ingresos
ventas_completo['Mes'] = ventas_completo['Fecha'].dt.month
ingresos_por_mes = ventas_completo.groupby('Mes')['Ingreso_Total'].sum()
mes_mayores_ingresos = ingresos_por_mes.idxmax()
print(f"Mes con mayores ingresos: {mes_mayores_ingresos}")
print("\n")
print("------------------------------------------------")
# 6. Calcula qu√© porcentaje de las ventas tuvo descuento
porcentaje_descuento = (ventas_completo['Descuento'] > 0).mean() * 100
print(f"Porcentaje de ventas con descuento: {porcentaje_descuento}%")
print("\n")
print("------------------------------------------------")
# 7. Crea un resumen ejecutivo con las m√©tricas m√°s importantes
resumen_ejecutivo = {
    'Ingreso_Total_Por_Producto': ingreso_total_producto,
    'Producto_Mas_Vendido': producto_mas_vendido,
    'Ticket_Promedio_Por_Tipo_Cliente': ticket_promedio_cliente,
    'Mes_Mayores_Ingresos': mes_mayores_ingresos,
    'Porcentaje_Ventas_Con_Descuento': porcentaje_descuento
}
print(resumen_ejecutivo)
````
