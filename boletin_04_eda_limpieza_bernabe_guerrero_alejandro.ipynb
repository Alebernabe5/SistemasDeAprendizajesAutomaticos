## 游리 Ejercicio 5: Visualizaci칩n Exploratoria - Histogramas y Boxplots

Crea el siguiente dataset de salarios y realiza visualizaciones:

```python
rng = np.random.default_rng(42)
# Salarios normales
salarios_normales = rng.normal(35000, 8000, 95)
# A침adimos algunos outliers (directivos)
salarios_outliers = np.array([100000, 120000, 150000, 200000, 250000])
salarios = np.concatenate([salarios_normales, salarios_outliers])

df_salarios = pd.DataFrame({'salario': salarios})
```

**Tareas:**
1. Crea un **histograma** de los salarios con 20 bins
2. A침ade l칤neas verticales para la media (roja) y la mediana (verde)
3. Crea un **boxplot** de los salarios
4. Identifica cu치ntos outliers hay usando el m칠todo IQR (Q1 - 1.5*IQR, Q3 + 1.5*IQR)
5. Muestra los valores de los outliers detectados

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Crear el DataFrame
rng = np.random.default_rng(42)
salarios_normales = rng.normal(35000, 8000, 95)
salarios_outliers = np.array([100000, 120000, 150000, 200000, 250000])
salarios = np.concatenate([salarios_normales, salarios_outliers])

df_salarios = pd.DataFrame({'salario': salarios})

# Preparacion de lineas par el grafico 

# Tu c칩digo aqu칤
# 1. Crea un **histograma** de los salarios con 20 bins
plt.figure(figsize=(10,5))
plt.hist(df_salarios['salario'], bins=20, color='skyblue', edgecolor='black')
# 2. A침ade l칤neas verticales para la media (roja) y la mediana (verde)
media = df_salarios['salario'].mean()
mediana = df_salarios['salario'].median()
plt.axvline(media, color='red', linestyle='--', label=f'Media: {media:.2f}')
plt.axvline(mediana, color='green', linestyle='--', label=f'Mediana: {mediana:.2f}')
plt.legend()
# 3. Crea un **boxplot** de los salarios
plt.figure(figsize=(10,4))
plt.boxplot(df_salarios['salario'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
# 4. Identifica cu치ntos outliers hay usando el m칠todo IQR (Q1 - 1.5*IQR, Q3 + 1.5*IQR)
Q1 = df_salarios['salario'].quantile(0.25)
Q3 = df_salarios['salario'].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
outliers = df_salarios[(df_salarios['salario'] < limite_inferior) | (df_salarios['salario'] > limite_superior)]
print("N칰mero de outliers:", len(outliers))
# 5. Muestra los valores de los outliers detectados
print("L칤mite superior detectado:", f"{limite_superior:.2f}")
print("Cantidad de  outliers detectados:", f"{len(outliers)}")
print("Valores detectados:", outliers['salario'].values)
````

## 游리 Ejercicio 7: An치lisis de Correlaciones

Crea un dataset con varias variables num칠ricas y analiza sus correlaciones:

```python
rng = np.random.default_rng(42)
n = 200

# Creamos variables con diferentes niveles de correlaci칩n
experiencia = rng.uniform(0, 20, n)
educacion = rng.integers(1, 5, n)  # 1=Secundaria, 2=Grado, 3=Master, 4=Doctorado
salario = 20000 + experiencia * 2000 + educacion * 5000 + rng.normal(0, 5000, n)
horas_trabajo = 40 + rng.normal(0, 5, n)
satisfaccion = 5 + 0.3 * experiencia - 0.1 * horas_trabajo + rng.normal(0, 1, n)
satisfaccion = np.clip(satisfaccion, 1, 10)

df_correlaciones = pd.DataFrame({
    'experiencia': experiencia,
    'educacion': educacion,
    'salario': salario,
    'horas_trabajo': horas_trabajo,
    'satisfaccion': satisfaccion
})
```

**Tareas:**
1. Calcula la matriz de correlaci칩n
2. Crea un **heatmap** de la matriz de correlaci칩n con `sns.heatmap()`
3. Identifica las dos variables con mayor correlaci칩n positiva
4. Identifica las dos variables con mayor correlaci칩n negativa
5. Crea un **scatter plot** entre experiencia y salario, a침adiendo una l칤nea de tendencia

````python
import seaborn as sns
import matplotlib.pyplot as plt
# Crear el DataFrame
rng = np.random.default_rng(42)
n = 200

experiencia = rng.uniform(0, 20, n)
educacion = rng.integers(1, 5, n)
salario = 20000 + experiencia * 2000 + educacion * 5000 + rng.normal(0, 5000, n)
horas_trabajo = 40 + rng.normal(0, 5, n)
satisfaccion = 5 + 0.3 * experiencia - 0.1 * horas_trabajo + rng.normal(0, 1, n)
satisfaccion = np.clip(satisfaccion, 1, 10)

df_correlaciones = pd.DataFrame({
    'experiencia': experiencia,
    'educacion': educacion,
    'salario': salario,
    'horas_trabajo': horas_trabajo,
    'satisfaccion': satisfaccion
})

# Tu c칩digo aqu칤

# 1. Calcula la matriz de correlaci칩n
matriz_correlacion = df_correlaciones.corr()
print("Matriz de correlaci칩n:", matriz_correlacion)
# 2. Crea un **heatmap** de la matriz de correlaci칩n con `sns.heatmap()`
plt.figure(figsize=(8,6))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', center=0)
plt.title('Matriz de Correlaci칩n')
plt.show()
# 3. Identifica las dos variables con mayor correlaci칩n positiva
series_corr = matriz_correlacion.unstack()
series_corr = series_corr[series_corr < 0.99]
pos_var, pos_valor = series_corr.idxmax(), series_corr.max()
print(f"3. Mayor correlaci칩n positiva: {pos_var[0]} y {pos_var[1]} (Valor: {pos_valor:.2f})")
# 4. Identifica las dos variables con mayor correlaci칩n negativa
neg_var, neg_valor = series_corr.idxmin(), series_corr.min()
print(f"4. Mayor correlaci칩n negativa: {neg_var[0]} y {neg_var[1]} (Valor: {neg_valor:.2f})")
# 5. Crea un **scatter plot** entre experiencia y salario, a침adiendo una l칤nea de tendencia
sns.regplot(x='experiencia', y='salario', data=df_correlaciones, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Experiencia vs Salario')
plt.xlabel('A침os de Experiencia')
plt.ylabel('Salario')
plt.show()
````

## 游리 Ejercicio 8: Tratamiento de Outliers

Trabaja con datos que contienen valores extremos:

```python
rng = np.random.default_rng(42)
# Datos con outliers intencionados
df_outliers = pd.DataFrame({
    'producto': ['P' + str(i) for i in range(1, 21)],
    'precio': [20, 25, 22, 28, 24, 500, 21, 26, -10, 23,
               27, 19, 25, 30, 22, 400, 24, 28, 26, -5],
    'unidades_vendidas': [100, 150, 120, 80, 200, 5, 130, 90, 110, 140,
                          95, 180, 160, 70, 145, 3, 125, 105, 155, 85]
})
```

**Tareas:**
1. Identifica outliers en 'precio' usando el m칠todo IQR
2. Identifica outliers l칩gicos (precios negativos o mayores a 100)
3. **Opci칩n A**: Elimina las filas con outliers
4. **Opci칩n B**: Reemplaza los outliers por los l칤mites (winsorizaci칩n)
5. **Opci칩n C**: Reemplaza los outliers por la mediana
6. Compara las estad칤sticas antes y despu칠s de cada tratamiento

````python
import pandas as pd
import numpy as np
# Crear el DataFrame
rng = np.random.default_rng(42)
df_outliers = pd.DataFrame({
    'producto': ['P' + str(i) for i in range(1, 21)],
    'precio': [20, 25, 22, 28, 24, 500, 21, 26, -10, 23,
               27, 19, 25, 30, 22, 400, 24, 28, 26, -5],
    'unidades_vendidas': [100, 150, 120, 80, 200, 5, 130, 90, 110, 140,
                          95, 180, 160, 70, 145, 3, 125, 105, 155, 85]
})

# Tu c칩digo aqu칤

# 1. Identifica outliers en 'precio' usando el m칠todo IQR
Q1 = df_outliers['precio'].quantile(0.25)
Q3 = df_outliers['precio'].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
outliers_iqr = df_outliers[(df_outliers['precio'] < limite_inferior) | (df_outliers['precio'] > limite_superior)]
print("Outliers detectados por IQR:", outliers_iqr)
# 2. Identifica outliers l칩gicos (precios negativos o mayores a 100)
outliers_logicos = df_outliers[(df_outliers['precio'] < 0) | (df_outliers['precio'] > 100)]
print("Outliers l칩gicos detectados:", outliers_logicos)

mediana_precio = df_outliers['precio'].median()

# 3. **Opci칩n A**: Elimina las filas con outliers
df_sin_outliers = df_outliers[~df_outliers.index.isin(outliers_iqr.index) & ~df_outliers.index.isin(outliers_logicos.index)]
print("DataFrame sin outliers:", df_sin_outliers)
# 4. **Opci칩n B**: Reemplaza los outliers por los l칤mites (winsorizaci칩n)
df_winsorizado = df_outliers.copy()
df_winsorizado['precio'] = df_winsorizado['precio'].clip(lower=limite_inferior, upper=limite_superior)
print("DataFrame winsorizado:", df_winsorizado)
# 5. **Opci칩n C**: Reemplaza los outliers por la mediana
df_mediana = df_outliers.copy()
df_mediana.loc[df_outliers['precio'].isin(outliers_iqr['precio']) | df_outliers['precio'].isin(outliers_logicos['precio']), 'precio'] = mediana_precio
print("DataFrame con mediana:", df_mediana)
# 6. Compara las estad칤sticas antes y despu칠s de cada tratamiento
print("Estad칤sticas originales:")
print(df_outliers['precio'].describe())
print("Estad칤sticas sin outliers:")
print(df_sin_outliers['precio'].describe())
````

df_sucio = pd.DataFrame({

**Tareas de Limpieza:**
1. **Duplicados**: Elimina clientes duplicados por 'cliente_id'
2. **Nombres**: Limpia espacios, estandariza formato (Nombre Apellido)
3. **Emails**: Normaliza a min칰sculas, identifica emails inv치lidos (sin '@' o dominio)
4. **Edad**: Identifica valores imposibles (negativos o >120), reemplaza con NaN
5. **Ciudad**: Estandariza nombres, corrige errores tipogr치ficos
6. **Ingresos**: Identifica outliers (negativos o extremadamente altos), decide tratamiento
7. **Fechas**: Unifica el formato de todas las fechas a 'YYYY-MM-DD'
8. **Reporte final**: Muestra un resumen del dataset limpio

````python
import pandas as pd
import numpy as np
# Crear el DataFrame
df_sucio = pd.DataFrame({
    'cliente_id': [1, 2, 3, 4, 5, 2, 6, 7, 8, 3, 9, 10],
    'nombre': ['  Ana Garc칤a  ', 'CARLOS RUIZ', 'mar칤a L칩pez', 'Pedro-S치nchez', 
               'laura MARTINEZ', 'CARLOS RUIZ', '  Juan P칠rez', 'Sofia_Gomez',
               'Miguel Torres', 'mar칤a L칩pez', 'Elena Castro', 'David  Moreno'],
    'email': ['ana@mail.com', 'carlos@MAIL.COM', 'maria@mail', 'pedro@mail.com',
              'laura@mail.com', 'carlos@MAIL.COM', 'juan@mail.com', None,
              'miguel@mail.com', 'maria@mail', 'elena@mail.com', 'david@mail.com'],
    'edad': [28, 35, 42, -5, 31, 35, 150, 29, 38, 42, 26, 45],
    'ciudad': ['Madrid', 'barcelona', 'VALENCIA', 'Madrid', 'Madird',
               'barcelona', 'Sevilla', 'Bilbao', 'Malaga', 'VALENCIA',
               'Madrid', 'barcelna'],
    'ingresos_mensuales': [2500, 3500, 4000, None, 2800, 3500, 3200, 2900,
                          -1000, 4000, 2600, 50000],
    'fecha_registro': ['2023-01-15', '2023/02/20', '15-03-2023', '2023-04-10',
                       '2023.05.25', '2023/02/20', '2023-06-30', '07/15/2023',
                       '2023-08-05', '15-03-2023', '2023-09-12', '2023-10-01']
})

print("=== DATASET ORIGINAL (SUCIO) ===")
print(df_sucio)
print(f"\nDimensiones: {df_sucio.shape}")

# Tu c칩digo de limpieza aqu칤

# 1. **Duplicados**: Elimina clientes duplicados por 'cliente_id'
df_limpio = df_sucio.drop_duplicates(subset=['cliente_id'], keep='first').copy()
# 2. **Nombres**: Limpia espacios, estandariza formato (Nombre Apellido)
df_limpio['nombre'] = df_limpio['nombre'].str.strip()  # Quitar espacios laterales
df_limpio['nombre'] = df_limpio['nombre'].str.replace(r'[-_]', ' ', regex=True) # Quitar guiones y barras
df_limpio['nombre'] = df_limpio['nombre'].str.title()  # Formato "Nombre Apellido"
# 3. **Emails**: Normaliza a min칰sculas, identifica emails inv치lidos (sin '@' o dominio)
df_limpio['email'] = df_limpio['email'].str.lower().str.strip()

mask_email_valido = df_limpio['email'].str.contains(r'[^@]+@[^@]+\.[a-z]{2,}', na=False)
df_limpio.loc[~mask_email_valido, 'email'] = np.nan
# 4. **Edad**: Identifica valores imposibles (negativos o >120), reemplaza con NaN
dicc_ciudades = {
    'barcelna': 'Barcelona', 'barcelona': 'Barcelona',
    'VALENCIA': 'Valencia', 'Madird': 'Madrid'
}
df_limpio['ciudad'] = df_limpio['ciudad'].replace(dicc_ciudades).str.capitalize()
````

````python

# Contin칰a la limpieza aqu칤
# 5. **Ciudad**: Estandariza nombres, corrige errores tipogr치ficos
df_limpio.loc[(df_limpio['edad'] < 0) | (df_limpio['edad'] > 120), 'edad'] = np.nan
df_limpio['edad'] = df_limpio['edad'].fillna(df_limpio['edad'].median())
# 6. **Ingresos**: Identifica outliers (negativos o extremadamente altos), decide tratamiento
df_limpio.loc[df_limpio['ingresos_mensuales'] < 0, 'ingresos_mensuales'] = np.nan
#
q_ingresos = df_limpio['ingresos_mensuales'].quantile(0.95)
df_limpio.loc[df_limpio['ingresos_mensuales'] > q_ingresos, 'ingresos_mensuales'] = df_limpio['ingresos_mensuales'].median()


# 7. **Fechas**: Unifica el formato de todas las fechas a 'YYYY-MM-DD'
df_limpio['fecha_registro'] = pd.to_datetime(df_limpio['fecha_registro'], dayfirst=False, errors='coerce')
````
