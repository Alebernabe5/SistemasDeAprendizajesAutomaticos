- F√≥rmula: `(x - min) / (max - min)`
- Los valores deben quedar entre 0 y 1

**M√©todo 2: Estandarizaci√≥n (Z-score)**
- F√≥rmula: `(x - media) / desviaci√≥n_est√°ndar`
- Los valores deben tener media 0 y desviaci√≥n est√°ndar 1

   - Media de los datos normalizados: deber√≠a ser aproximadamente 0 (ej: 0.0000001)
   - Desviaci√≥n est√°ndar: deber√≠a ser aproximadamente 1 (ej: 0.9999 o 1.0001)
